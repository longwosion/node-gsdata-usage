```python
%matplotlib inline
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties 
font = FontProperties(fname=r"/Library/Fonts/Microsoft/SimHei.ttf", size=16)  
styles = ['-1', '-d', '-x', '-+', '-s', '-o', '-v', '-<', '->', '-^', 
          '--1', '--d', '--x', '--+', '--s', '--o', '--v', '--<', '-->', '--^']
```

```python
data = pd.read_csv('wci.data.txt')
data = data.set_index('date')
```

```python
ax = data.plot(kind='line',figsize=(32, 12), fontsize=16, grid=True, style=styles, color='black', legend='reverse')

labels = ax.legend(bbox_to_anchor=(1.12, 1.01)).texts
for label in labels : 
    label.set_fontproperties(font) 
plt.xlabel(u'数据统计时间', fontsize=16).set_fontproperties(font)
plt.ylabel(u'微信传播力指数').set_fontproperties(font)
plt.show()
```

```python
data_raw = pd.read_csv('url_num2.txt')
data2 = data_raw.set_index('date')
ax = data2.plot(kind='line',figsize=(20, 20), fontsize=15, grid=True, style=styles, color='black', legend='reverse')

labels = ax.legend(loc='best').texts
font = FontProperties(fname=r"/Library/Fonts/Microsoft/SimHei.ttf", size=15)  
for label in labels : 
    label.set_fontproperties(font) 
plt.xlabel(u'数据统计时间', fontsize=16).set_fontproperties(font)
plt.ylabel(u'各微信号发表文章数累计').set_fontproperties(font)
plt.show()
```

```python
ax_bar = data_raw.set_index('date')[-1:].sum()[0:].plot(kind='barh', figsize=(8, 10), color='gray')

labels = ax_bar.get_yticklabels()
for label in labels : 
    label.set_fontproperties(font) 
    
font = FontProperties(fname=r"/Library/Fonts/Microsoft/SimHei.ttf", size=12)      
plt.xlabel(u'50周各微信号发表文章数累计').set_fontproperties(font)
plt.show()
```
